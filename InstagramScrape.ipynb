```python
session_id = "38566737751%3Ah7JpgePGAoLxJe%334" #instagram's session ID 
profiles = result
output = pd.DataFrame()
for id in profiles:
    post = InstagramPost(id, sessionid=session_id)
    #print("Post author:",post.author) #Just printted the results for debugging purposes. 
    #print("Post caption:",post.text)
    #print("Number of likes of post:",post.number_of_likes)
    #print('-----'*20)
    data = {'Author': post.author,
        'likes': post.number_of_likes,
        'caption': post.text,
        'date': post.upload_time}
    
    output = output.append(data, ignore_index=True)
```

### This code is using the InstagramPost class from the instagramy library to scrape Instagram post information.

- The session_id variable is set to a string containing a session ID for the Instagram API.

- The profiles variable is set to a list of Instagram post IDs.

- The output variable is initialized as an empty DataFrame.

- The code then uses a for loop to iterate over each post ID in the profiles list.

- For each post ID, it creates a new InstagramPost object by passing the post ID and session ID to the constructor of the class.
 
- Then it creates a dictionary with four keys: 'Author', 'likes', 'caption' and 'date' and assigns the values of the properties 'author', 'number_of_likes', 'text', and 'upload_time' from the InstagramPost object to the corresponding keys respectively.

- Finally, it appends this dictionary to the DataFrame by using the append() method and passing the data dictionary to it, and also passing the ignore_index parameter as True. This will add the data into the DataFrame without preserving the original index and creates a new index for each row.
